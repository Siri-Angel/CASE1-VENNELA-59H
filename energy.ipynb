```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
mouna=pd.read_csv('energy.csv')
print(mouna)
srh=mouna[mouna['State']=='Andhra']
rcb=mouna[mouna['State']=='Karnataka']
fig1,diag1=plt.subplots(2)
diag1[0].plot(srh['Date'],srh['Consumption'])
diag1[1].plot(rcb['Date'],rcb['Consumption'],color='red')

#barc  chart telengana tripura
revanth=mouna[mouna['State']=='Telengana']
cm=mouna[mouna['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(revanth['Date'],revanth['Consumption'],color='red')
diag[1].bar(cm['Date'],cm['Consumption'],color='yellow')

State1='Delhi'
State2='Odisha'
State3='Up'
State4='Punjab'
cse5=[State1,State2]

cse4=mouna[mouna['State']==State1]['Consumption'].sum()
cse3=mouna[mouna['State']==State2]['Consumption'].sum()
cse2=[cse3,cse4]
fig,diag=plt.subplots(2)
diag[0].pie(cse2,labels=cse5,autopct='%1.1f%%')
cse5=[State1,State2]
cse4=mouna[mouna['State']==State1]['Consumption'].sum()
cse3=mouna[mouna['State']==State2]['Consumption'].sum()
cse2=[cse3,cse4]
diag[1].pie(cse2,labels=cse5,autopct='%1.1f%%')
```
